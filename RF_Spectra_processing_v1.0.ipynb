```python
#@title

!git clone https://github.com/cavalcante-lucas/RF_Spectra_processing
from sklearn.preprocessing import StandardScaler
import joblib
import markdown
from ipywidgets import widgets, HBox, VBox, Dropdown, HTML
from IPython.display import display, clear_output
import pandas as pd
import sys
from io import StringIO
import numpy as np
import matplotlib.pyplot as plt
import io
%matplotlib inline
from google.colab import files

#########################%##############%########################

#datasets
reference_T = pd.read_csv('archives/dataset_trabalho_T.csv')
reference_T = reference_T.drop('Unnamed: 0',axis=1)
reference_S = pd.read_csv('archives/dataset_trabalho_S.csv')
reference_S = reference_S.drop('Unnamed: 0',axis=1)
reference_columns = pd.read_csv('archives/puras_selecionadas.csv')
reference_columns = reference_columns.drop('Unnamed: 0',axis=1)

#########################%##############%########################

#lists
pred = []
data = []

#########################%##############%########################

#fuctions
def scalling (reference,data):
    scaler = StandardScaler()
    reference_params = scaler.fit(reference)
        
    data = df.add_prefix('X')
    data = pd.DataFrame(data, columns=reference.columns)
    data = data.reset_index().drop('index',axis=1)
    data_scaled = reference_params.transform(data)
    data_scaled = pd.DataFrame(data_scaled,columns=reference.columns)
    
    return(data_scaled)

####################################

def load_predict(model, data):
    loaded_rf = joblib.load(model)
    predictions = loaded_rf.predict(data)
    
    for i in range(len(predictions)):
        if predictions[i] == 1:
            pred.append('Adulterated')
        
        elif predictions[i] == 0:
            pred.append('pure')
    
    pred_final = pd.DataFrame(pred, columns=['Model Prediction'])
    pred_final.index.names = ['Sample']
    
    return(pred_final)

####################################
    
def load_spectra(a):
    global df
    
    if data_format.value == 'CSV':
        for i in range(len(list(uploader.value.values()))):
            input_file = list(uploader.value.values())[i]
            content = input_file['content']
            content = io.StringIO(content.decode('utf-8'))
            ints = pd.read_csv(content)['Intensity']
            data.append(ints)

        df = pd.DataFrame(data)
        df.columns = reference_columns.columns

    else:
        for i in range(len(list(uploader.value.values()))):
            input_file = list(uploader.value.values())[i]
            content = input_file['content']
            content = io.StringIO(content.decode('utf-8'))
            ints = pd.read_csv(content)
            data.append(ints)

        df = pd.DataFrame(np.array(data).ravel()).transpose()
        df = pd.DataFrame(df.values.reshape(3, -1))
        df = df.drop([0,1,2,3,4],axis=1).reset_index(drop=True)
        df.columns = reference_columns.columns
    
    return(df)

####################################

def model_results(a):
    global pred_final
    
    if analyte_select.value == 'Tadalafil':
        data_scaled = scalling(reference_T,df)

        model = 'archives/models/RF_1.0Hold_9833(00276)CV_MACA_T_s101.joblib'
        pred_final = load_predict(model, data_scaled)
        print(pred_final)

    else:
        data_scaled = scalling(reference_S,df)

        model = 'archives/models/RF_1.0Hold_9917(00250)CV_MACA_S_s0.joblib'
        pred_final = load_predict(model, data_scaled)
        print(pred_final)
    
    return(pred_final)

#########################%##############%########################

#widgets

analyte_select =widgets.ToggleButtons(
    options=['Tadalafil', 'Sildenafil'],
    description='Analyte:',
    disabled=False,
    button_style='',
    tooltips=['Select the Analyte', 'Select the Analyte'],
)

data_format = widgets.RadioButtons(
    options=['CSV', 'ASP'],
    description='Data format:',
    disabled=False
)

uploader = widgets.FileUpload(accept='*.csv, *.asp', multiple=True)

botao_predict = widgets.Button(
    value=False,
    description='Predict',
    disabled=False,
    button_style='success',
    tooltip='Predict the data',
    icon='check' 
)

botao_predict.on_click(load_spectra)
botao_predict.on_click(model_results)

vertical1 = widgets.VBox([analyte_select,data_format])

horizontal = widgets.HBox([vertical1,uploader])

horizontal2 = widgets.HBox([horizontal,botao_predict])

accordion = widgets.Accordion(children=[horizontal2], titles=('Predict Data'))


display(accordion)
```
